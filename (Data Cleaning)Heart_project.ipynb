```python
df['Checkup']=df['Checkup'].replace({'Within the past 2 years':'<=2 years',
                                     'Within the past year':'<=1 year',
                                    '5 or more years ago':'>=5 years',
                                    'Within the past 5 years':'<=5 years'})
```

```python
#converting Alcohol_consumption, Fruit_consumption, Green_Vegetables_Consumption and FriedPotato_Consumption into 'yes/no'
```

```python
def convert_to_yes_no(value):
    return 'Yes' if value != 0 else 'No'

df['Alcohol_Consumption']=df['Alcohol_Consumption'].apply(convert_to_yes_no)
df['Fruit_Consumption']=df['Fruit_Consumption'].apply(convert_to_yes_no)
df['Green_Vegetables_Consumption']=df['Green_Vegetables_Consumption'].apply(convert_to_yes_no)
df['FriedPotato_Consumption']=df['FriedPotato_Consumption'].apply(convert_to_yes_no)
```

```python
#MNaming the index 'Patients' as there are no other patient identification columns
```

```python
df.index = df.index +1
df = df.rename_axis("Patient")
df.head(3)
```
